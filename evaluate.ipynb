```python
# names = [
#     "C:/Users/Kuma/PycharmProjects/Eventgo_exp/dqn_batch_size_5/pertrain_dqn_6_choose_evaluationdata_fixstep=3",
#          ]
names = [#'pretrain_model/pretrain_multi_google_rank_transformer_choose.h5_evaluationdata_fixstep=3',
         # 'pretrain_model/pretrain_multi_google_rank_transformer_tstep5_news_choose.h5_evaluationdata_fixstep=3',
# 'pretrain_model/pretrain_multi_google_rank_transformer_pick_30_choose.h5_evaluationdata_fixstep=3'
# 'pretrain_model/pretrain_dqn_new_step6_buffer_size10k_epsilon_c_choose.h5_evaluationdata_fixstep=6',
    #'dqn_pre_model/pretrain_dqn_new_step6_epoch_70_buffer_size10k_bertbisltm_choose.h5_evaluationdata_fixstep=6',
    #'fixstepmodel/SARSA_step3_coor_bertbisltm_top3_fixreward_choose.h5_evaluationdata',
    #'finetunemodel/Actor_finetune_A2C_nofixstep_eplision_0.3_epoch1.h5_evaluationdata',
    #'stepnofixmodel/pretrain_dqn_train40_epoch100_choose.h5_evaluationdata',
    'stepnofixmodel/pretrain_dqn_new_nofixstep_bertbisltm_coor_top3_cost0.4_fixreward_choose.h5_evaluationdata_asset_max300_del_click_same'
]
ROIs = []
Recalls = []
UnitCost = []
Nums = []

for name in names:
    convertthreshold = False
    newthreshold = 0
    anslen = []
    for i in range(0,1):
        dynamicdata = dict(Tool.LoadObj(name))
        newthreshold = 0.1+(i*0.1)
        print(newthreshold)

        positivetotalDF = pd.read_csv('dataset_google_search/newdata_test - remove error url.csv')
        import csv
        with open('negativapage.csv', newline='',encoding='utf-8') as f:
            reader = csv.reader(f)
            negativeapge = list(reader)
            #print(p)
            negativepage = [','.join(p[0:-1]) for p in negativeapge if ':' in ','.join(p[0:-1])]

        # positivepage = list(positivetotalDF['Event Source Page URL'])
        positivepage = list(positivetotalDF['Event source page'])
        # positivepage = URLTool.GetNormalizaUrl(positivepage)
        ans = []

        print("positivepage len",len(positivepage))
        with open('positivepage.csv', newline='',encoding='utf-8') as f:
            reader = csv.reader(f)
            tmp = list(reader)
            positivepage += [','.join(p[0:-1]) for p in tmp if ':' in ','.join(p[0:-1])]
        # for id,p in enumerate(positivepage):
        #     try:
        #         p = p[p.index(":"):]
        #     except Exception:
        #         print('error',id)
        positivepage = [p[p.index(":"):] for p in positivepage]
        negativepage = [p[p.index(":"):] for p in negativepage]
        recall = 0
        total = 0
        totalsteps = 0
        nonezerowebsite = 0
        test = 0
        print("add after positivepage len",len(positivepage))

        for i,hrefs in enumerate(list(dynamicdata['ClickHref'])):
            flag = False
            totalsteps += len(hrefs)
            if len(hrefs)>0:
                nonezerowebsite+=1
            for j,h in enumerate(hrefs):
            # for j,h in enumerate(hrefs):
                #print(dynamicdata['ClickProb'][i][j])
                if float(dynamicdata['ClickProb'][i][j])>newthreshold or not convertthreshold:
                    try:
                        h = h[h.index(':'):]
                    except Exception as e:
                        print(e)
                        continue
                    if h in positivepage:
                        test+=1
                        ans.append(1)
                        flag = True
                    elif h in negativepage:
                        ans.append(0)
                    else:
                        ans.append(0)
                        print("step:{},ClickHomepage:{}, Click AT:{}, Click Href:{}".format(j,dynamicdata['ClickHomepage'][i],dynamicdata['ClickAT'][i][j],dynamicdata['ClickHref'][i][j]))
                else:
                    print('continue')
                    break
            anslen.append(test)
            if flag:
                recall += 1
            total += 1
        print(newthreshold)
        print(nonezerowebsite)
        ROIs.append(sum(ans)/len(ans))
        UnitCost.append(totalsteps/sum(ans))
        Recalls.append(recall/total)
        print('ROI:',sum(ans)/len(ans))
        print('UnitCost',totalsteps/sum(ans))
        print('ESP',sum(ans))
        print('of total click',len(ans))
        print('episodes reaching ESP',recall)
        print('# number of episodes',total)
        print('total cost',totalsteps)
        print('ESP',sum(ans))
        print('Recall:',recall/total)
        print('F1',2*(sum(ans)/len(ans))*(recall/total)/((sum(ans)/len(ans))+(recall/total)))
        print('avg.step',totalsteps/total)
        Nums.append(sum(ans))
```
